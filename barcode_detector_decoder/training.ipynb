```python
import zipfile
import os
from ultralytics import YOLO

# âœ… Step 1: Extract the dataset
zip_path = "/content/BC_nuevo_externo.v5-without-augmentation.yolov8.zip"
extract_path = "/content/BC_nuevo_externo.v5-without-augmentation.yolov8"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("âœ… Dataset extracted successfully!")

# âœ… Step 2: Define the dataset path
data_yaml_path ="/content/BC_nuevo_externo.v5-without-augmentation.yolov8/BC_nuevo_externo.v5-without-augmentation.yolov8/data.yaml"

# âœ… Step 3: Load YOLOv8 model
model = YOLO("yolov8n.pt")  # Using YOLOv8 nano model (change to yolov8s.pt, yolov8m.pt if needed)

# âœ… Step 4: Train the model
model.train(data=data_yaml_path, epochs=50, imgsz=640, batch=16)

print("ðŸš€ Training Complete!")

```
